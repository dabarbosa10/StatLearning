```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,1,100)
y=np.sin(x*2*np.pi)+0.13*np.random.normal(size=x.shape)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y,'o', markerfacecolor="None")
plt.plot(x,np.sin(x*2*np.pi),'g-')
plt.show()
```

```python
xran=np.sort(np.random.randint(0,100,15))
yran=y[xran]
x1=x[xran]
```

```python
plt.figure(figsize=(10,6))
plt.plot(x1,yran,'o', markerfacecolor="None")
plt.xlabel('x')
plt.ylabel('t')
plt.show()
```

```python
m=np.array([0,1,3,9])
```

```python
def DesgMatrix(samples,order):
    feat=samples.shape[0]
    phi=np.ones((feat,order))
    for i in range(1,order):
        phi[:,i]=phi[:,i]*(samples**i)
    return phi
```

```python
X3=DesgMatrix(x1,2)
X5=DesgMatrix(x1,5)
X10=DesgMatrix(x1,11)
```

```python
beta3=np.matmul(np.matmul(np.linalg.pinv(np.matmul(X3.T,X3)),X3.T),yran)
beta5=np.matmul(np.matmul(np.linalg.pinv(np.matmul(X5.T,X5)),X5.T),yran)
beta10=np.matmul(np.matmul(np.linalg.pinv(np.matmul(X10.T,X10)),X10.T),yran)
```

```python
X31=DesgMatrix(x,2)
plt.figure(figsize=(10,6))
plt.plot(x1,yran,'o', markerfacecolor="None")
plt.plot(x,np.matmul(X31,beta3),'r-')
plt.plot(x,np.sin(np.pi*2*x),'g-')
plt.ylim(-1.5,1.5)
plt.xlabel('x')
plt.ylabel('t')
plt.show()
```

```python
X51=DesgMatrix(x,5)
plt.figure(figsize=(10,6))
plt.plot(x1,yran,'o', markerfacecolor="None")
plt.plot(x,np.matmul(X51,beta5),'r-')
plt.plot(x,np.sin(np.pi*2*x),'g-')
plt.ylim(-1.5,1.5)
plt.xlabel('x')
plt.ylabel('t')
plt.show()
```
